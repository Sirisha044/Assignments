```python
!pip install -q transformers

from transformers import pipeline

generator = pipeline("text-generation", model="gpt2")

prompt = "Explain the concept of black holes in simple terms."

output = generator(prompt, max_new_tokens=50)
print(output)
```

```python
!pip install -q transformers

from transformers import pipeline

# Use task-specific translation model
translator = pipeline("translation_en_to_fr", model="Helsinki-NLP/opus-mt-en-fr")

# 1. Real Translation
text = "Hello, how are you?"
output = translator(text, max_length=40)[0]['translation_text']
print("🔹 Zero-shot Translation (REAL):\n", output)

# Use FLAN for other tasks
flan = pipeline("text2text-generation", model="google/flan-t5-large")

def run(prompt, max_tokens=60):
    return flan(prompt, max_new_tokens=max_tokens)[0]["generated_text"].strip()

# 2. Summarization
print("\n🔹 Summarization:\n", run("Summarize in one line: The sun is the closest star to Earth and is the center of the solar system."))

# 3. Question Answering
print("\n🔹 Question Answering:\n", run("What is the capital of France? Only give the name."))

# 4. Few-shot Translation with FLAN
prompt = """Translate English to Spanish:
English: Good morning
Spanish: Buenos días
English: Thank you
Spanish:"""
print("\n🔹 Few-shot Translation:\n", run(prompt))

# 5. Chain-of-Thought
prompt = "Ram had 8 apples and gave 3 to Bob. How many apples does Ram have now? Answer only the number."
print("\n🔹 Chain-of-Thought Reasoning:\n", run(prompt))
```
